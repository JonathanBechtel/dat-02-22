```python
###Questions
##How do we scrape embedded tags? e.g. the tag for average rating is 'div,' but it's embedded in 'span'


#I only split the tags from the first page. How do I loop through a nested list 
#and generate a new variable for each element without manually creating new variables?
#Neighborhood stopped working when I was able to loop through each page 


from bs4 import BeautifulSoup
import pandas as pd
import requests


#there are 24 pages 
page_ranges = range(10, 240, 10)
restaurants = []
phone_numbers = []
addresses = []
neighborhood = []
num_reviews = []
categories = []
containers = []
url = []
req = []
scraper = []


#first page first. First page base URL is different from subsequent URLs
    
url = "https://www.yelp.com/search?find_desc=Restaurants&find_loc=London%2C%20United%20Kingdom"
req = requests.get(url)
scraper = BeautifulSoup(req.text)
containers += scraper.find_all('div', {'class': 'container__09f24__21w3G'})
restaurants += [container.find_all('a')[1].text for container in containers]
phone_numbers += [container.find_all('p')[1].text for container in containers]
addresses += [container.find_all('p')[2].text for container in containers]
#neighborhood += [container.find_all('p')[3].text for container in containers]
num_reviews += [container.find_all('span')[4].text for container in containers]
categories += [container.find_all('span')[7].text for container in containers]
    
#subsequent pages: for each i, or number, in page_ranges, insert i into the url and repeat this loop for each url 

for i in page_ranges:
    url = "https://www.yelp.com/search?find_desc=Restaurants&find_loc=London%2C%20United%20Kingdom&start=" + str(i)
    req = requests.get(url)
    scraper = BeautifulSoup(req.text)
    containers += scraper.find_all('div', {'class': 'container__09f24__21w3G'})
    restaurants += [container.find_all('a')[1].text for container in containers]
    phone_numbers += [container.find_all('p')[1].text for container in containers]
    addresses += [container.find_all('p')[2].text for container in containers]
    #neighborhood += [container.find_all('p')[3].text for container in containers]
    num_reviews += [container.find_all('span')[4].text for container in containers]
    categories += [container.find_all('span')[7].text for container in containers]
    


```

```python
#split the tags and create indicator variables true/false for tags
tags = [i.split(',') for i in categories]

    
fish_chips = False
indian = False
british = False
steakhouse = False
french = False
coffee_tea = False
breakfast_brunch = False
american = False
soul_food = False
cajun_creole = False
italian = False
modern_euro = False
cocktail = False
    
for i in tags:
    for j in tags:
        if 'Fish & Chips' in j:
            fish_chips = True
        if 'Indian' in j:
            indian = True
        if 'British' in j:
            british = True
        if 'Steakhouses' in j:
            steakhouse = True
        if 'Coffee & Tea' in j:
            coffee_tea = True
        if ' Breakfast & Brunch' in j:
            breakfast_brunch = True
        if 'American (Traditional)' in j:
            american = True
        if ' Soul Food' in j:
            soul_food = True 
        if ' Cajun/Creole' in j:
            cajun_crole = True
        if 'Italian' in j:
            italian = True
        if ' Modern European' in j:
            modern_euro = True
        if ' Cocktail Bars' in j:
            cocktail = True
            
```
